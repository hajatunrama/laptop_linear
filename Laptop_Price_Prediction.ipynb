## 10.2 Identifikasi Tipe Data

Feature dibagi menjadi dua tipe agar dapat diproses dengan metode preprocessing yang tepat.

````python
categorical_cols = X.select_dtypes(include="object").columns.tolist()
numeric_cols     = X.select_dtypes(exclude="object").columns.tolist()

print("="*40)
print(f"Categorical Features ({len(categorical_cols)} kolom):")
print("="*40)
for col in categorical_cols:
    print(f"  - {col}")

print()

print("="*40)
print(f"Numerical Features ({len(numeric_cols)} kolom):")
print("="*40)
for col in numeric_cols:
    print(f"  - {col}")
````

## 10.3 Analisis Cardinality

Cardinality menunjukkan jumlah kategori unik pada setiap feature kategorikal.
Informasi ini penting untuk menentukan metode encoding yang tepat.

````python
cardinality = pd.DataFrame({
    "Feature": categorical_cols,
    "Jumlah Kategori": [X[col].nunique() for col in categorical_cols],
    "Contoh Nilai": [str(list(df[col].unique()[:3])) for col in categorical_cols]
})

cardinality
````

Parameter `handle_unknown="ignore"` ditambahkan agar model tetap bisa berjalan jika ada nilai kategori yang tidak dikenali saat deployment.

## 10.4 Pembagian Data Latih dan Data Uji

Dataset dibagi dengan rasio **80:20** menggunakan `train_test_split` dengan `random_state=42` agar hasil dapat direproduksi.

````python
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=42
)

print("="*50)
print(f"Training Data   : {X_train.shape[0]:,} baris")
print(f"Testing Data    : {X_test.shape[0]:,} baris")
print(f"Rasio Train/Test: 80% / 20%")
print("="*50)
````

### Interpretasi

Sebanyak **80%** data (~964.072 baris) digunakan untuk pelatihan dan **20%** (~241.018 baris) untuk pengujian.

Strategi ini memastikan model dapat belajar dari data yang cukup besar, sekaligus memiliki data independen untuk evaluasi kemampuan generalisasi model.

## 10.5 ColumnTransformer — Preprocessing Pipeline

`ColumnTransformer` menggabungkan preprocessing untuk fitur kategorikal dan numerik menjadi satu langkah otomatis yang konsisten antara data latih dan data uji.

````python
preprocessor = ColumnTransformer(
    transformers=[
        (
            "cat",
            OneHotEncoder(handle_unknown="ignore", sparse_output=False),
            categorical_cols
        )
    ],
    remainder="passthrough"  # Fitur numerik dilewatkan apa adanya
)

print("✅ ColumnTransformer siap.")
print(f"  - Categorical: {categorical_cols}")
print(f"  - Numerical (passthrough): {numeric_cols}")
````

### Interpretasi

`ColumnTransformer` akan:
- Menerapkan **OneHotEncoder** pada semua feature kategorikal → mengubah menjadi kolom biner.
- Melewatkan semua feature numerik tanpa perubahan (`remainder="passthrough"`).

Karena diintegrasikan dalam `Pipeline`, transformasi ini hanya di-*fit* pada data latih dan di-*transform* pada data uji, mencegah data leakage.

# 11. 🤖 Model Development

Empat algoritma regresi dibangun menggunakan **Pipeline** yang menggabungkan preprocessing dan model dalam satu alur kerja.

Setiap model dievaluasi menggunakan MAE, RMSE, R² Score, dan waktu komputasi untuk memungkinkan perbandingan yang adil dan objektif.

````python
results = []
pipelines = {}

def evaluate_model(model_name, model):
    """
    Membuat pipeline, melatih model, melakukan prediksi,
    dan merekam semua metrik evaluasi.

    Parameters
    ----------
    model_name : str
        Nama model (untuk label di tabel hasil).
    model : sklearn estimator
        Objek model Machine Learning.

    Returns
    -------
    pipeline : sklearn Pipeline
        Pipeline yang sudah dilatih (fitted).
    """
    # Buat pipeline
    pipeline = Pipeline([
        ("preprocessor", preprocessor),
        ("model", model)
    ])

    # Pelatihan
    start_train = time.time()
    pipeline.fit(X_train, y_train)
    training_time = time.time() - start_train

    # Prediksi
    start_predict = time.time()
    y_pred = pipeline.predict(X_test)
    prediction_time = time.time() - start_predict

    # Metrik
    mae  = mean_absolute_error(y_test, y_pred)
    rmse = np.sqrt(mean_squared_error(y_test, y_pred))
    r2   = r2_score(y_test, y_pred)

    results.append({
        "Model"               : model_name,
        "MAE"                 : round(mae, 4),
        "RMSE"                : round(rmse, 4),
        "R2"                  : round(r2, 6),
        "Training Time (s)"   : round(training_time, 2),
        "Prediction Time (s)" : round(prediction_time, 4),
    })

    pipelines[model_name] = pipeline

    print(f"✅ [{model_name}] selesai | R²={r2:.4f} | MAE=${mae:,.2f} | RMSE=${rmse:,.2f} | Train={training_time:.1f}s")

    return pipeline

print("✅ Fungsi evaluate_model siap.")
````

## 11.1 Linear Regression

Linear Regression adalah **baseline model** — algoritma paling sederhana yang memodelkan hubungan linear antara feature dan target.

Digunakan sebagai patokan performa minimum yang harus dilampaui algoritma lain.

````python
linear_pipeline = evaluate_model(
    "Linear Regression",
    LinearRegression()
)
````

### Interpretasi

Linear Regression mengasumsikan hubungan linear antara fitur dan target. Model ini sangat cepat dilatih namun terbatas dalam menangkap pola non-linear.

Pada dataset ini, karena `Total_Performance` memiliki korelasi hampir linear terhadap `Price_USD`, Linear Regression diperkirakan sudah memberikan hasil yang sangat baik.

## 11.2 Decision Tree Regressor

Decision Tree membangun struktur pohon keputusan yang membagi data berdasarkan kondisi fitur secara rekursif.

Mampu menangkap hubungan non-linear namun rentan terhadap **overfitting** jika tidak dibatasi kedalamannya.

````python
dt_pipeline = evaluate_model(
    "Decision Tree",
    DecisionTreeRegressor(
        random_state=42
    )
)
````

### Interpretasi

Decision Tree tanpa batasan kedalaman (`max_depth=None`) memungkinkan pohon tumbuh sampai setiap daun murni — ini berpotensi menyebabkan overfitting pada data latih.

Perbandingan antara performa pada data latih dan data uji penting untuk diperhatikan pada model ini.

## 11.3 Random Forest Regressor

Random Forest adalah metode **ensemble** yang membangun banyak Decision Tree secara paralel menggunakan subset data dan fitur yang berbeda, lalu mengambil rata-rata prediksi.

Lebih stabil dari Decision Tree tunggal karena mengurangi varians melalui rata-rata banyak pohon.

````python
rf_pipeline = evaluate_model(
    "Random Forest",
    RandomForestRegressor(
        n_estimators=30,  # Dikurangi agar tidak terlalu lambat pada 1.2M data
        random_state=42,
        n_jobs=-1         # Gunakan semua core CPU
    )
)
````

### Interpretasi

# 13. 🔁 Cross-Validation (5-Fold)

Cross-validation memberikan evaluasi yang lebih robust dibandingkan satu kali train-test split.

Dengan **5-Fold Cross-Validation**, data dibagi menjadi 5 bagian — model dilatih 5 kali, setiap kali menggunakan 4 bagian untuk latih dan 1 bagian untuk validasi. Rata-rata dari 5 skor memberikan estimasi generalisasi yang lebih andal.

> **Catatan**: Karena dataset berukuran 1.2 juta baris, cross-validation akan memakan waktu cukup lama. Cross-validation dilakukan pada **subsample 50.000 baris** agar tetap representatif namun efisien.

````python
# Subsample untuk efisiensi CV
from sklearn.utils import resample

X_cv, y_cv = resample(
    X, y,
    n_samples=50000,
    random_state=42
)

cv_results = {}

models_for_cv = {
    "Linear Regression" : LinearRegression(),
    "Decision Tree"      : DecisionTreeRegressor(random_state=42),
    "Random Forest"      : RandomForestRegressor(n_estimators=30, random_state=42, n_jobs=-1),
    "XGBoost"            : XGBRegressor(random_state=42, n_jobs=-1, verbosity=0),
}

print("Cross-Validation (5-Fold, n=50.000) sedang berjalan...")
print("="*55)

for name, model in models_for_cv.items():
    pipe_cv = Pipeline([
        ("preprocessor", preprocessor),
        ("model", model)
    ])
    start = time.time()
    scores = cross_val_score(
        pipe_cv,
        X_cv,
        y_cv,
        cv=5,
        scoring="r2",
        n_jobs=-1
    )
    elapsed = time.time() - start
    cv_results[name] = scores
    print(f"{name:22} | Mean R²={scores.mean():.4f} | Std={scores.std():.4f} | {elapsed:.1f}s")

print("="*55)
print("✅ Cross-Validation selesai.")
````

````python
# Visualisasi CV
cv_df = pd.DataFrame(cv_results)

plt.figure(figsize=(10, 5))
cv_df.boxplot(column=list(cv_results.keys()), grid=False)
plt.ylabel("R² Score")
plt.title("Cross-Validation R² Score (5-Fold)", fontsize=13)
plt.xticks(rotation=15)
plt.tight_layout()
plt.show()

# Tabel ringkasan
cv_summary = pd.DataFrame({
    "Model"     : list(cv_results.keys()),
    "CV Mean R²": [v.mean().round(4) for v in cv_results.values()],
    "CV Std R²" : [v.std().round(4) for v in cv_results.values()],
})
cv_summary
````

### Interpretasi

Cross-validation memverifikasi bahwa performa model konsisten di berbagai subset data, bukan hanya kebetulan pada satu split tertentu.

- **Mean R² tinggi + Std rendah** = model stabil dan generalisasi baik.
- **Std tinggi** = model tidak konsisten, perlu tuning lebih lanjut.

Model dengan CV Mean R² tertinggi dan Std terendah adalah kandidat terkuat untuk dipilih sebagai model final.
